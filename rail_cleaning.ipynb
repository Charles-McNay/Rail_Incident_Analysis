### First I will import the libraries I need to clean my data

```python
import pandas as pd
import numpy as np
```

### I will use the read_csv function to read in my data so I'm not changing the original file

```python
df = pd.read_csv('rr_data.csv', low_memory=False)
df.head(10)
```

### I have some columns that are unnecessary to answer my questions for this project, here I'm using the drop() function and inplace=True to remove those columns on the original dataframe

```python
df.drop(['Class', 'Hours Engineers On Duty', 'Maintenance Railroad Name', 'Equipment Attended'], axis=1, inplace=True)
```

### I'm using the info() function to inspect my datafram for data types, row and column counts and to see how many NaN's I have to start

```python
df.info()
```

### Getting a list of columns as they're written

```python
df.columns
```

### I want to rename the columns to remove spaces and unnecessary words to make working with the data easier, here I use the rename() function on a dictionary to rename the columns

```python
df.rename(columns= {'Equipment Type Code': 'Equipment', 
'Weather Condition': 'Weather', 
'Train Speed': 'Speed', 
'Reporting Railroad Name': 'Railroad', 
'State Name': 'State', 
'Equipment Damage Cost': 'Damage',
'Primary Accident Cause': 'Cause',
'Remote Control Locomotive Code': 'RCO',
'Passengers Transported': 'Passengers',
'Accident Type': 'Accident_Type',
'Persons Evacuated': 'Persons_Evacuated',
'Positive Alcohol Tests': 'Alcohol_Test',
'Positive Drug Tests': 'Drug_Test',
'Hazmat Released Cars': 'Hazmat_Spill_Cars',
'Total Persons Killed': 'Lives_Lost',
'Gross Tonnage': 'Tonnage'}, inplace=True)

```

### This project is focused on freight trains and my dataset contains passenger trains. This dataset is pulled from the FRA's incident form and on that form I have a key to codes for passenger trains. Here I'm using the drop() function to remove any rowes that are related to passenger trains.

```python
df.drop(df.loc[df['Equipment']=='2'].index, inplace=True)
df.drop(df.loc[df['Equipment']=='3'].index, inplace=True)
df.drop(df.loc[df['Equipment']=='B'].index, inplace=True)
df.drop(df.loc[df['Equipment']=='C'].index, inplace=True)
df.drop(df.loc[df['Equipment']=='EMU'].index, inplace=True)
df.drop(df.loc[df['Equipment']=='DMU'].index, inplace=True)
df
```

### I no longer need the Equipment column and will delete it using the drop() function.

```python
df = df.drop('Equipment', axis=1)
```

### Using the info() function to check data types again

```python
df.info()
```

### From looking over the source for the data, if the Railroad name is not listed the row is a error so I am removing rows that have a NaN in those columns.

```python
df.dropna(subset=['Railroad'], inplace=True)
```

### Replacing NaN values with the appropriate values using fillna()

```python
df.Weather = df.Weather.fillna('Unknown')
df.Cause = df.Cause.fillna('Unknown')
df.Passengers = df.Passengers.fillna(0)
df.Accident_Type = df.Accident_Type.fillna('Unknown')
df.RCO = df.RCO.fillna(0)
df.Speed = df.Speed.fillna(0)
df.Alcohol_Test = df.Alcohol_Test.fillna(0)
df.Drug_Test = df.Drug_Test.fillna(0)
```

```python
df.info()
```

### After dropping rows the index needs to be reset so I'm using the reset_index() function to ensure the rows are numbered correctly moving forward

```python
df = df.reset_index(drop=True)
df.index
```

### The data types need to be changed to make analysis easier, I'm using the astype() function with a dictonary to assign the correct dtypes

```python
df = df.astype({'Tonnage': 'Int64', 'Damage': 'Int64', 
'Lives_Lost': 'Int64', 'Drug_Test': 'Int64', 'Speed': 'Int64',
'Alcohol_Test': 'Int64', 'Hazmat_Spill_Cars': 'Int64',
'Persons_Evacuated': 'Int64'})
df['Date'] = df['Date'].astype('datetime64[ns]')
df.info()
```
